#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
#url ='https://github.com/trending/python?since=daily'
url='https://en.wikipedia.org/wiki/Python'
````

````python
#your code
html4 = requests.get(url).content
soup4 = BeautifulSoup(html4, "lxml")
print(soup4)
````

````python
links = soup4.find_all("li")
links2 = [element.a["href"] for element in links if element.a != None]
links2
````

````python
links_list = []
for link in links2:
    if link.startswith("#"):
        links_list.append("https://en.wikipedia.org/wiki/Python"+link)
    elif link.startswith("/wiki/"):
        links_list.append("https://en.wikipedia.org"+link)
    else:
        links_list.append(link)
links_list
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
html7 = requests.get(url).content
soup7 = BeautifulSoup(html7, "lxml")
print(soup7)
````

````python
earthquakes1 = [element.text.replace('\xa0',',').strip().split("\n") for element in soup7.find_all("tr", {"class":"ligne1 normal"})] 
earthquakes2 = [element.text.replace('\xa0',',').strip().split("\n") for element in soup7.find_all("tr", {"class":"ligne2 normal"})]
earthquakes1
earthquakes2
````

````python
earthquakes = earthquakes1 + earthquakes2
earthquakes
````

````python
data = [functools.reduce(lambda a,b,c,d,e : a+","+b+","+c+","+d+","+e, element) for element in earthquakes]
data
````

````python
df=pd.DataFrame(data)
df.head(5)
````

#### Display the date, days, title, city, country of next 25 hackathon events as a Pandas dataframe table

````python
# This is the url you will scrape in this exercise
#url ='https://hackevents.co/hackathons'
url='https://mlh.io/seasons/na-2020/events'
````

````python
#your code
html8 = requests.get(url).content
soup8 = BeautifulSoup(html8, "lxml")
print(soup8)
````

````python
titlename=[name.text.replace('\n\n','\n').strip().split("\n") for name in soup8.find_all('h3',{'class':'event-name'})]
titlename

class_dates=soup8.find_all('p',{'class':'event-date'})
dates=[date.text.replace('\n\n','\n').strip().split("\n") for date in soup8.find_all('p',{'class':'event-date'})]
dates

class_locations=soup8.find_all('div',{'class':'event-location'})
locations=[location.text.replace('\n\n','\n').strip().split("\n") for location in soup8.find_all('div',{'class':'event-location'})]
locations
````

````python
#Data Frame

df1=pd.DataFrame(titlename)
df2=pd.DataFrame(dates)
df3=pd.DataFrame(locations)
````

````python
#Concatenate Dataframes

df1= pd.concat([df1, df2], axis=1,)
df3= pd.concat([df1, df3], axis=1,)
df3.columns = ['Name','Date','Place','Status']
df3
````
